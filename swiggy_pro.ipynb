```python
#encoding
Nominal_cols=['type_of_vehicle','festival','city_name','is_weekend']
Ordinal_cols=['weather','traffic']
Numerical_cols=['age','ratings','vehicle_condition','multiple_deliveries','pickup_time_minutes',
                'order_time_hour','distance']
```

```python
order_ = [['sunny','cloudy','fog','windy','sandstorms','stormy'],['low','medium','high','jam']]
```

```python
df['traffic'].unique()
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import OneHotEncoder,OrdinalEncoder,StandardScaler
from sklearn.linear_model import LinearRegression
```

```python
Transformer=ColumnTransformer([('One Hot Encoding',OneHotEncoder(),Nominal_cols),
                  ('Ordinal Encoding',OrdinalEncoder(),Ordinal_cols),
                   ('Feature Scaling',StandardScaler(),Numerical_cols)])
```

```python
Transformer
```

```python
#Model Building
#splitting data into input and output
X = df.drop('time_taken',axis = 1)
y = df['time_taken']
```

```python
#splitting data into training and testing
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=24)
```

```python
X_train.shape
```

```python
y_train.shape
```

```python
X_test.shape
```

```python
y_test.shape
```

```python
model_lr = Pipeline([('Data Preprocessing',Transformer),
          ("LinearRegression",LinearRegression())])
model_lr
```

```python
from sklearn.metrics import r2_score,mean_squared_error
```

```python
model_lr.fit(X_train,y_train)
y_pred = model_lr.predict(X_test)
print("MSE",mean_squared_error(y_test,y_pred))
print("r2 Score",r2_score(y_test,y_pred))
```

```python
#KNN Regression
from sklearn.neighbors import KNeighborsRegressor
model_knn = Pipeline([('Data Preprocessing',Transformer),
          ("LinearRegression",KNeighborsRegressor())])
model_knn.fit(X_train,y_train)
y_pred = model_knn.predict(X_test)
print("MSE",mean_squared_error(y_test,y_pred))
print("r2 Score",r2_score(y_test,y_pred))
```

```python
#Supportvector Machine
from sklearn.svm import SVR
model_svr = Pipeline([('Data Preprocessing',Transformer),
          ("LinearRegression",SVR())])
model_svr.fit(X_train,y_train)
y_pred = model_svr.predict(X_test)
print("MSE",mean_squared_error(y_test,y_pred))
print("r2 Score",r2_score(y_test,y_pred))
```

```python
#Decision tree
from sklearn.tree import DecisionTreeRegressor
model_dt = Pipeline([('Data Preprocessing',Transformer),
          ("LinearRegression",DecisionTreeRegressor())])
model_dt.fit(X_train,y_train)
y_pred = model_dt.predict(X_test)
print("MSE",mean_squared_error(y_test,y_pred))
print("r2 Score",r2_score(y_test,y_pred))
```

```python
#Random Forest
from sklearn.ensemble import RandomForestRegressor
model_rf = Pipeline([('Data Preprocessing',Transformer),
          ("LinearRegression",RandomForestRegressor())])
model_rf.fit(X_train,y_train)
y_pred = model_rf.predict(X_test)
print("MSE",mean_squared_error(y_test,y_pred))
print("r2 Score",r2_score(y_test,y_pred))
```
